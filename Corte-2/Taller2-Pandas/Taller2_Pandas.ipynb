# <b> PUNTO 1 - CREACION Y EXPLORACION DEL DATAFRAME <b/>

## Importacion de libreria

En este caso, llamaremos a las liberias con las cuales vamos a trabajar a lo largo del trabajo. Usaremos Pandas Y Numpy para trabajar con el analisis de datos.



```python
import pandas as pd
import numpy as np
```

# Crear el DataFrame

El paso a seguir es usar las funciones de pandas para crear el DataFrame con datos propocionados desde un archivo CSV.

```python
URL = "https://raw.githubusercontent.com/EdwarSuarezQ/Curso-Inteligencia-Artificial/refs/heads/main/Corte-2/Taller2-Pandas/productos.csv"

df = pd.read_csv(URL, sep=";")

print(df)
```

## Mostrar las primeras 5 filas.

Con Pandas, podemos visualizar las primeras filas o registros de un DataFrame con la funcion **.head()**, despues le indicamos una cantidad, en este caso 5.

```python
primeros = df.head(5)

print(primeros)
```

## Mostrar las ultimas 10 filas

En caso de que queramos ver las últimas 10 filas o registros de un DataFrame usamos **.tail().**

```python
ultimos = df.tail(10)

print(ultimos)
```

## Mostrar el tipo de dato de cada columna

Con la funcion **.dtypes** podemos ver el tipo de datos de cada columna.

```python
tipoDatos = df.dtypes

print(tipoDatos)
```

## Mostrar el tamaño de un DataFrame.

La funcion **.shape** nos permite concer el tamaño exacto de un DataFrame con sus filas y columna.

```python
filas, colmnas = df.shape

print(f"La cantidad de filas es: {filas}. La cantidad de columnas es: {colmnas}.")
```

## Generar una descripción estadística del conjunto de los datos.

Para generar una descripcion del conjunto de datos usamos **.describe()** y obtener una estadistica general de DataFrame.


```python
descripcionEstadistica = df.describe()

print(descripcionEstadistica)
```

# <b> Punto 2 – Operaciones Matemáticas entre Columnas </b>

En esta seccion crearemos nuevas culumnas y usaremos operaciones matematicas


## Crear una nueva columna llamada Total_Venta

Para calcular el total de venta de cada producto multiplicamos la columna **Precio** por la columna **Cantidad**.

```python
df["Total_Venta"] = df["Precio"] * df["Cantidad"]

print(df[["Precio", "Total_Venta"]].head(10))
```

## Crear una nueva columna llamada Costo_Total

Para calcular el costo total sumamos la columna **Total_Venta** con la columna **Costo_Envio**.

```python
df["Costo_Total"] = df["Total_Venta"] + df["Costo_Envio"]

print(df[["Producto", "Costo_Total"]].head(10))
```

## Calcular el promedio del Total_Venta

Con la funcion **.mean()** podemos calcular el promedio de una columna.

```python
promedio = df["Total_Venta"].mean()

print(f"El promedio del Total_Venta es: {promedio}")
```

## Identificar el producto con mayor Total_Venta

Con **.idxmax()** podemos obtener la posicion del valor mas alto de una columna.

```python
mayor = df.loc[df["Total_Venta"].idxmax()]

print("El producto con mayor Total_Venta es:")
print(mayor)
```

## Identificar el producto con menor Total_Venta

Con **.idxmin()** podemos obtener la posicion del valor más bajo de una columna.

```python
menor = df.loc[df["Total_Venta"].idxmin()]

print("El producto con menor Total_Venta es:")
print(menor)
```

#  Punto 3 - Porcentajes y Análisis

## Crear columna llamada IVA

Para calcular el precio del IVA multiplicamos la columna **Total_venta** x el 0.19, que corresponde al 19%.

```python
df["IVA"] = df["Total_Venta"] * 0.19

print(df[["Producto", "IVA"]].head(10))
```

## Crear una columna llamada Ganancia_Estimada

Para calcular la ganacia estimada multiplicamos **Total_Venta** por 0.25, que corresponde al 25%

```python
df["Ganancia_Estimada"] = df["Total_Venta"] * 0.25

print(df[["Producto", "Ganancia_Estimada"]].head(10))
```

## Crear una columna llamada Perdida_Estimada

Para calcular la perdida estimada multiplicamos **Total_Venta** por 0.0005, correspondiente al 0.05%.

```python
df["Perdida_Estimada"] = df["Total_Venta"] * 0.0005

print(df[["Producto", "Perdida_Estimada"]].head(10))
```

## Calcular el porcentaje de ventas de cada producto

Para calcular el porcentaje de ventas primero obtenemos el total general y luego calculamos que porcentaje representa cada producto.

```python
total_general = df["Total_Venta"].sum()

df["Porcentaje_Ventas"] = (df["Total_Venta"] / total_general) * 100

print(df[["Producto", "Porcentaje_Ventas"]].head(10))
```

## Mostrar productos con porcentaje superior al 15%

Con una condición podemos filtrar únicamente los productos cuyo porcentaje de ventas sea mayor al 15%.

```python
venta_superiores = df[df["Porcentaje_Ventas"] > 15]

print(venta_superiores[["Producto", "Porcentaje_Ventas"]])
```

# Punto 4 - Estadística Descriptiva

En esta caso usaremos las funciones de Pandas para hacer estadisticas descriptivas.

## Calcular la media, mediana y moda de la columna Precio

Con las funciones **.mean(), .median() y .mode()** podemos obtener diferentes medidas estadísticas de una columna numérica.

```python
media = df["Precio"].mean()
mediana = df["Precio"].median()
moda = df["Precio"].mode()

print(f"Media: {media}")
print(f"Mediana: {mediana}")
print(f"Moda: {moda}")
```

## Calcular la desviación estándar de la columna Cantidad

Con **.std()** podemos calcular quu tan dispersos estan los valores respecto al promedio.

```python
desviacion_estandar = df["Cantidad"].std()

print(f"Desviacion estandar: {desviacion_estandar}")
```

## Calcular el valor máximo y mínimo de Total_Venta

Con **.max() y .min()** podemos obtener el valor mas alto y bajo de una columna.

```python
maximo = df["Total_Venta"].max()
minimo = df["Total_Venta"].min()

print(f"Valor maximo: {maximo}")
print(f"Valor minimo: {minimo}")
```

# Punto 5 - Filtrado

## Mostrar unicamente los productos de la categoria Tecnologia

Con una condicional podemos filtrar únicamente los productos que pertenezcan a la categoría Tecnología.

```python
tecnologia = df[df["Categoria"] == "Tecnología"]

print(tecnologia[["Producto", "Categoria"]].head(10))
```

## Ordenar los productos de mayor a menor según Total_Venta

Con **.sort_values()** podemos ordenar los valores de una columna. En este caso ordenamos **Total_Venta** de mayor a menor.

```python
ordenar = df.sort_values(by="Total_Venta", ascending=False)

print(ordenar[["Producto", "Total_Venta"]].head(10))
```
