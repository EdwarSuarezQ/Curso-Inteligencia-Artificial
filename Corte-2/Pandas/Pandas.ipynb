## ¿Que es pandas?

Es una libreria de codigo abierto para programar en Python.
Proporciona estructuras de datos y herramientas para analizar datos de alto rendimiento y son muy faciles de usar. Su nombre proviene de la expresion "Panel Data".

## ¿Para que sirve?

Principamente para la manipulacion, exploracion, limpieza y analisis de datos. Permita trabaja con datos estructurados (tablas, hojas de cálculo, bases de datos SQL, archivos CSV, etc.) de forma intuitiva, lo que facilita operaciones como filtrado, agrupación, combinación, manejo de valores nulos y transformación de datos en general.

## Objetivo principal de la librería

El objetivo principal de Pandas es proporcionar los bloques de construcción fundamentales de alto nivel necesarios para hacer el análisis práctico de datos reales en Python rápido y fácil. Busca ser la herramienta más potente y flexible para la manipulación de datos open-source disponible en cualquier lenguaje.

## Principales funciones y características

-   **Estructuras de datos:** Introduce los DataFrame (estructuras 2D tabulares) y las Series (arrays 1D indexados).

-   **Manejo de datos faltantes:** Herramientas eficientes para detectar, eliminar o rellenar valores nulos (NaN).

-   **Operaciones vectorizadas:** Al estar construida sobre NumPy, las operaciones numéricas sobre las columnas son extremadamente rápidas sin necesidad de usar bucles **for** explícitos.

-   **Alineación inteligente de datos:** Evita errores comunes al asegurar que los datos estén alineados según sus etiquetas (índices) durante los cálculos.

-   **Agrupación y agregación:** Funcionalidad poderosa ***groupby*** para dividir, aplicar y combinar conjuntos de datos.

-   **Lectura/Escritura versátil:** Soporta una amplia variedad de formatos (CSV, Excel, SQL, JSON, etc.).

-   **Fusión (Merge/Join):** Herramientas fáciles de usar para cruzar y unir múltiples conjuntos de dato.

## 1. Importar libreria
Primero importamos las librerias necesarias para trabaja, en este caso **Pandas Y Numpy** .

```python
import pandas as pd
import numpy as np
```

## 2. Creamos un DataFrame.

Un DataFrame es una estructura de datos bidimensional y tabular(compuestas por filas y columnas), similar a una hoja de calculo en Excel o una tabla en SQL.

En este caso, crearemos un DataFrame a partir de un diccionario, que tendra las nostas de 5 estudiantes en tres cortes, las notas seran evaludas de 0-100

```python
datos_alumnos = {
    "Estudiante": ["Miguel", "Sofía", "Javier", "Laura", "Andrés"],
    "Corte 1": [90, 64, 56, 100, 70],
    "Corte 2": [80, 76, 60, 96, 84],
    "Corte 3": [76, 80, 70, 90, 80]
}


df_notas = pd.DataFrame(datos_alumnos)

#Para visualizar el DataFrame
print("Registro de Notas inicial:")
display(df_notas)
```

## 3. Operaciones entre columnas.

Tomaremos las notas definivas de cada corte para saber el promedio de cada estudiante.

```python
df_notas["Promedio Actual"] = ((df_notas["Corte 1"] + df_notas["Corte 2"] + df_notas["Corte 3"]) / 3).round(1)

print("\nDataFrame con el cálculo del promedio:")
display(df_notas)
```

## 5. Datos sinteticos.

Son informacion generada artificialmente mediante algoritmos y modelos de inteligencia artificial que la imitan las caracteristicas estadisticas, relaciones y estructuras de datos reales.

```python
np.random.seed(123) # Para reproducibilidad

# Generamos notas aleatorias para 10 alumnos en cada uno de los 3 cortes
datos_sinteticos = {
    "ID_Alumno": range(1001, 1011),
    "Corte 1": np.random.randint(30, 101, 10),
    "Corte 2": np.random.randint(30, 101, 10),
    "Corte 3": np.random.randint(30, 101, 10)
}

df_sintetico = pd.DataFrame(datos_sinteticos)

# Reutilizamos la lógica del paso 3 para sacar su promedio final
df_sintetico["Promedio Final"] = ((df_sintetico["Corte 1"] +
                                   df_sintetico["Corte 2"] +
                                   df_sintetico["Corte 3"]) / 3).round(1)

print("\nDatos sintéticos generados:")
display(df_sintetico)
```

## 6. Estadisticas basicas.

Con Pandas es facil obtener un resumen estadistico descriptivo (media, maximos, minimos, desviaciones).

```python
#El metodo .describe() nos dara la estadistica de cada corte y del promedio final
print("\nEstadísticas descriptivas generales de la clase sintética:")
display(df_sintetico.describe())
```

## 7. Filtrar datos



```python
aprobados = df_sintetico[df_sintetico["Promedio Final"] >= 60]

print(f"\nDe los 10 estudiantes simulados, se encontraron {len(aprobados)} estudiantes aprobados.")
print("Muestra de los aprobados:")
display(aprobados)
```
