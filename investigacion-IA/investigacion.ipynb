# Algoritmo A* (A Star)

El algoritmo A* es un algoritmo de busqueda que es utilizado para encontrar el camino mas corto entre dos nodos dentro de un grafo.

Es uno de los algoritmos mas utilizados en inteligencia artificial para resolver problemas de navegacion, planificacion de rutas y videojuegos.

El algoritmo combina el costo real del camino recorrido con una estimacion del costo restante para llegar al objetivo.

## Teoria del algoritmo A*

El algoritmo A* utiliza una funcion de evaluacion para determinar cual nodo explorar primero.

La formula utilizada es:

f(n) = g(n) + h(n)

donde:

g(n) = costo real desde el nodo inicial hasta el nodo actual  
h(n) = estimacion del costo desde el nodo actual hasta el objetivo (heuristica)  
f(n) = costo total estimado del camino

El algoritmo siempre selecciona el nodo con el menor valor de f(n) para continuar la busqueda.

## Como funciona el algoritmo

El algoritmo A* funciona de la siguiente manera:

1. Se define un nodo inicial.
2. Se agregan los nodos vecinos a una lista de exploracion.
3. Se calcula el valor f(n) para cada nodo.
4. Se selecciona el nodo con menor costo estimado.
5. El proceso continua hasta encontrar el nodo objetivo.

Gracias al uso de una heuristica, el algoritmo puede encontrar soluciones de manera mas eficiente que otros algoritmos de busqueda.

## Grafo utilizado

Para el ejemplo se utilizara el siguiente grafo:

A -> B (1)  
A -> C (4)  

B -> D (2)  
B -> E (5)  

C -> F (3)  

D -> G (1)  
E -> G (2)  
F -> G (5)

El objetivo sera encontrar el camino mas corto desde el nodo A hasta el nodo G.

```python
import heapq

grafo = {
    'A': {'B': 1, 'C': 4},
    'B': {'D': 2, 'E': 5},
    'C': {'F': 3},
    'D': {'G': 1},
    'E': {'G': 2},
    'F': {'G': 5},
    'G': {}
}

heuristica = {
    'A': 7,
    'B': 6,
    'C': 2,
    'D': 1,
    'E': 2,
    'F': 3,
    'G': 0
}

def a_star(grafo, inicio, objetivo):

    cola = []
    heapq.heappush(cola, (0, inicio))

    costos = {inicio: 0}
    padres = {inicio: None}

    while cola:

        _, actual = heapq.heappop(cola)

        if actual == objetivo:
            break

        for vecino, costo in grafo[actual].items():

            nuevo_costo = costos[actual] + costo

            if vecino not in costos or nuevo_costo < costos[vecino]:

                costos[vecino] = nuevo_costo
                prioridad = nuevo_costo + heuristica[vecino]

                heapq.heappush(cola, (prioridad, vecino))
                padres[vecino] = actual

    camino = []
    nodo = objetivo

    while nodo is not None:
        camino.append(nodo)
        nodo = padres[nodo]

    camino.reverse()

    return camino


inicio = "A"
objetivo = "G"

camino = a_star(grafo, inicio, objetivo)

print("Nodo inicial:", inicio)
print("Nodo objetivo:", objetivo)
print()
print("Camino encontrado:")
print(" -> ".join(camino))
```

## Resultados:


Al ejecutar el algoritmo A* se obtiene el siguiente camino:

A -> B -> D -> G

Esto significa que el algoritmo encontro el camino mas corto entre el nodo inicial y el nodo objetivo.

# Heuristica

## ¿Que es?

Una heuristica es una funcion o metodo que permite estimar el costo o la distancia para llegar a un objetivo dentro de un problema de busqueda.

En inteligencia artificial, las heuristicas se utilizan para guiar los algoritmos hacia soluciones mas eficientes, reduciendo el numero de caminos que deben explorarse.

En lugar de analizar todas las posibles soluciones, una heuristica ayuda a tomar decisiones mas rapidas basadas en una estimacion.

## ¿Para que sirven?

Las heuristicas sirven para mejorar la eficiencia de los algoritmos de busqueda.

Entre sus principales funciones se encuentran:

- Reducir el numero de nodos que se deben explorar
- Guiar la busqueda hacia las soluciones mas prometedoras
- Disminuir el tiempo de ejecucion de los algoritmos
- Facilitar la resolucion de problemas complejos

Muchos algoritmos de inteligencia artificial utilizan heuristicas, como el algoritmo A*.

## ¿Como funciona?

Una heuristica funciona estimando el costo restante para llegar a un objetivo.

Esta estimacion permite que el algoritmo tome decisiones mas inteligentes sobre cual camino explorar primero.

Por ejemplo, en el algoritmo A* la heuristica representa una estimacion de la distancia entre el nodo actual y el nodo objetivo.

Gracias a esta informacion, el algoritmo puede priorizar los caminos que parecen mas cercanos a la solucion.

## Ejemplo 1:

Un ejemplo comun de heuristica es el uso de la distancia en linea recta en los sistemas de navegacion GPS.

Cuando una aplicacion calcula una ruta, puede estimar la distancia directa entre dos puntos para determinar cual camino es mas corto.

Esta estimacion ayuda al sistema a encontrar rutas eficientes mas rapidamente.

## Ejemplo 2

En los videojuegos, los personajes controlados por inteligencia artificial utilizan heuristicas para encontrar el camino hacia un objetivo.

Por ejemplo, un enemigo puede calcular cuantos pasos necesita para llegar a la posicion del jugador.

Con esta informacion el personaje puede elegir el camino mas corto o mas rapido.

# Teoria de juegos

## ¿Que es?

La teoria de juegos es una rama de las matematicas y de la economia que estudia las decisiones que toman varios participantes cuando las decisiones de uno afectan a los demas.

En este tipo de situaciones, cada participante busca obtener el mayor beneficio posible teniendo en cuenta las posibles acciones de los otros jugadores.

La teoria de juegos se utiliza en inteligencia artificial, economia, politica, biologia y en el analisis de estrategias.

## ¿Para que es?

La teoria de juegos sirve para analizar situaciones donde varias personas o agentes toman decisiones que influyen en los resultados de los demas.

Entre sus aplicaciones se encuentran:

- Analizar estrategias en economia
- Estudiar negociaciones y conflictos
- Desarrollar inteligencia artificial en videojuegos
- Analizar comportamientos competitivos

Gracias a la teoria de juegos es posible entender mejor como las decisiones individuales afectan a un sistema completo.

## ¿Como funciona?

La teoria de juegos funciona analizando las posibles estrategias que cada jugador puede elegir dentro de una situacion.

Cada jugador intenta escoger la estrategia que le permita obtener el mejor resultado posible.

El resultado final depende de la combinacion de decisiones tomadas por todos los jugadores.

En muchos casos se busca encontrar un equilibrio en el cual ningun jugador pueda mejorar su resultado cambiando su estrategia de manera individual.

## Ejemplo 1: Dilema del prisionero

El dilema del prisionero es uno de los ejemplos mas conocidos de la teoria de juegos.

Dos personas son arrestadas y tienen dos opciones: cooperar entre ellas o traicionarse.

Si ambos cooperan reciben una pena menor, pero si uno traiciona al otro puede obtener un beneficio mayor mientras el otro recibe una pena mas fuerte.

Este ejemplo muestra como las decisiones individuales pueden afectar el resultado de todos los participantes.


## Ejemplo 2: Ajedrez

El ajedrez es un ejemplo clasico de teoria de juegos.

Cada jugador debe analizar las posibles jugadas de su oponente antes de tomar una decision.

El objetivo es elegir movimientos que aumenten las probabilidades de ganar mientras se intenta anticipar las estrategias del adversario.


## Ejemplo 3: Subastas

En una subasta, varios participantes compiten por obtener un producto o servicio realizando ofertas de dinero. Cada participante debe decidir cuanto esta dispuesto a ofrecer teniendo en cuenta su propio interes y lo que cree que los demas participantes pueden ofrecer.

La estrategia de cada participante consiste en elegir una oferta que le permita ganar la subasta sin pagar mas dinero del necesario. Sin embargo, si la oferta es demasiado baja, otro participante podria ganar.

Por esta razon, las decisiones de cada participante dependen de las posibles decisiones de los demas, lo que convierte a las subastas en un escenario tipico de analisis en la teoria de juegos.
